```python
sectors=dict({1:"Basic Materials", 2:"Consumer Cyclical", 3:"Financial Services", 4:"Real Estate", 5:"Consumer Defensive", 6:"Healthcare", 7:"Utilities", 8:"Communication Services", 9:"Energy", 10:"Industrials", 11:"Technology"})
```

```python
def symbolsearch(l_symbols):
    output=dict()
    for symbol in l_symbols:
        url = "https://finance.yahoo.com/quote/"+symbol+"/profile?p="+symbol
        response = requests.get(url)

        if response.status_code == 200:
            pass
        else:
            print(symbol)
            raise AttributeError("Failed")
        
        results_page = BeautifulSoup(response.content,'lxml')
        
        try:
            tag=results_page.find('span',{'class':"Fw(600)"}).get_text()
            if len(tag)>23:
                output[symbol]=None
            else:
                output[symbol]=tag
        except AttributeError:
            try:
                url2 = "https://finance.yahoo.com/quote/"+symbol+"/holdings?p="+symbol
                response2 = requests.get(url2)

                if response2.status_code == 200:
                    pass
                else:
                    print(symbol)
                    raise TypeError("Failed")
            
                results_page2 = BeautifulSoup(response2.content,'lxml')
            
                l=dict()
                for i in range(1,12):
                    link=results_page2.find_all('span',{'class':"W(20%) D(b) Fl(start) Ta(e)"})[i]
                    per=link.get_text()[:-1]
                    l[i]=float(per)
                
                if all(l.values())==0.0:
                    output[symbol]=None
                else:
                    max_key = max(l, key=l.get)
                    output[symbol]=sectors[max_key]
            except AttributeError:
                output[symbol]=None
            except IndexError:
                output[symbol]=None
            except ValueError:
                output[symbol]=None
              
    return output
```
